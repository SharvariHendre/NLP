```python
# Cell 1: Imports and Setup
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import normalize
# from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

# Sample corpus
corpus = [
    "The quick brown fox jumps over the lazy dog.",
    "Never jump over the lazy dog quickly.",
    "Bright foxes are quick and clever.",
    "Dogs are loyal and friendly animals."
]
```

```python
# Cell 2: Bag-of-Words (Raw Count)
count_vectorizer = CountVectorizer()
bow_counts = count_vectorizer.fit_transform(corpus)
bow_df = pd.DataFrame(bow_counts.toarray(), columns=count_vectorizer.get_feature_names_out())
print("Bag-of-Words (Raw Count):")
display(bow_df)
```

```python
# Cell 3: Bag-of-Words (Normalized Count)
# Normalize each row (document) to unit norm
bow_normalized = normalize(bow_counts, norm='l1', axis=1)
bow_norm_df = pd.DataFrame(bow_normalized.toarray(), columns=count_vectorizer.get_feature_names_out())
print("Bag-of-Words (Normalized Count):")
display(bow_norm_df)
```

```python
# Cell 4: TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print("TF-IDF:")
display(tfidf_df)
```

```python
# Cell 5: Tokenization for Word2Vec
tokenized_corpus = [word_tokenize(doc.lower()) for doc in corpus]
print("Tokenized Corpus for Word2Vec:")
print(tokenized_corpus)
```

```python
! pip install --upgrade gensim
from gensim.models import Word2Vec
# Cell 6: Word2Vec Embedding
w2v_model = Word2Vec(sentences=tokenized_corpus, vector_size=100, window=5, min_count=1, workers=4, epochs=50)

# Example: show vector for a word
word = 'dog'
if word in w2v_model.wv:
    print(f"Vector for '{word}':\n", w2v_model.wv[word])
else:
    print(f"'{word}' not in vocabulary.")
```
